```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import japanize_matplotlib 
```

```python
tmp = pd.DataFrame({"rate": [1/ (math.log2(i) + 1) for i in range(1, 21)]})
```

```python
tmp["cumsum"] = tmp["rate"].cumsum()
```

```python
plt.ylim(0, 1.2)
plt.xlabel('予測ランキングでの順位')
plt.ylabel('割引率')
plt.plot(range(1, 21), tmp["rate"])
```

```python
plt.ylim(0, 7)
plt.xlabel('予測ランキングでの順位')
plt.ylabel('割引率累積')
plt.plot(range(1, 21), tmp["cumsum"])
```
