```python
import umap
from sklearn.datasets import load_digits
from scipy.sparse.csgraph import connected_components
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.manifold import TSNE
import time

import os

import numpy as np
import pandas as pd

```

```python
IMD_DIR = os.path.join(os.getcwd(), "../../data/intermediate/")

IMD_INPUT_DIR = os.path.join(IMD_DIR, "pickled_inputs")
```

```python
digits = load_digits()
digits.target = [float(digits.target[i]) for i in range(len(digits.target))]

# UMAP
start_time = time.time()
embedding = umap.UMAP().fit_transform(digits.data)
interval = time.time() - start_time
plt.scatter(embedding[:,0],embedding[:,1],c=digits.target,cmap=cm.tab10)
plt.colorbar()
```

```python
watch = pd.read_pickle(os.path.join(IMD_INPUT_DIR , "watch.pkl"))
auction = pd.read_pickle(os.path.join(IMD_INPUT_DIR , "auction.pkl"))
genre = pd.read_csv(os.path.join(os.getcwd(), "../../data/input/", "genre.csv"))
```

```python
watch = watch.merge(auction, on="AuctionID").merge(genre, on="GenreID", how="left")
```

```python
watch.columns
```

```python
tmp = pd.crosstab(watch['KaiinID'], watch['BrandID'])
```

```python
tmp.shape
```

```python
# UMAP
start_time = time.time()
embedding = umap.UMAP().fit_transform(np.array(pd.crosstab(watch['BrandID'], watch['KaiinID'][:20000])))
interval = time.time() - start_time
plt.scatter(embedding[:,0],embedding[:,1])
plt.colorbar()
```

```python
trans = umap.UMAP(n_neighbors=5,n_components=2, random_state=42)\
    .fit(np.array(pd.crosstab(watch['BrandID'], watch['KaiinID'][:20000])))
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
watch.shape
```

```python
model = LatentDirichletAllocation(n_components=10)
model.fit(np.array(pd.crosstab(watch['BrandID'], watch['KaiinID'][:100000])))
lda_out = pd.DataFrame(model.transform(np.array(pd.crosstab(watch['BrandID'], watch['KaiinID'][:100000]))))
lda_out
```

```python
# UMAP
start_time = time.time()
embedding = umap.UMAP().fit_transform(np.array(lda_out))
interval = time.time() - start_time
plt.scatter(embedding[:,0],embedding[:,1])
plt.colorbar()
```

```python
model = LatentDirichletAllocation(n_components=5)
model.fit(np.array(pd.crosstab(watch['KaiinID'], watch['BrandID'][:50000])))
lda_out = pd.DataFrame(model.transform(np.array(pd.crosstab(watch['KaiinID'], watch['BrandID'][:50000]))))
lda_out
```

```python
# UMAP
start_time = time.time()
embedding = umap.UMAP().fit_transform(np.array(lda_out))
interval = time.time() - start_time
plt.scatter(embedding[:,0],embedding[:,1])
plt.colorbar()
```
