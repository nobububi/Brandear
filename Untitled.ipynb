```python
from multiprocessing import Pool

import pandas as pd
import numpy as np
from catboost import CatBoost
from catboost import Pool
import lightgbm as lgb
```

```python
sushi_mst = pd.read_csv("input/sushi3.idata", sep="\t", header=None,
                       names=["item_id", "名前", "スタイル", "大分類", "小分類", "こってり度", "食べる頻度", "正規化した価格", "販売頻度"]).drop("名前", axis=1)
user_mst = pd.read_csv("input/sushi3.udata", sep="\t", header=None,
                       names=["user_id", "性別", "年齢", "回答時間", "長く住んだ県", "長く住んだ地域", "長く住んだ東西",
                              "現在の県", "現在の地域", "現在の東西", "移住"])
pref_melted = pd.read_csv("input/sushi3b.5000.10.score", sep="\s", header=None)
```

```python
sushi_mst
```

```python
user_mst
```

```python
pref = pref_melted.stack().reset_index().rename(columns={"level_0": "user_id", "level_1": "item_id", 0: "score"})
pref
```

```python
train_df = (
    user_mst.iloc[:4000, :].merge(pref, on="user_id", how="inner").merge(sushi_mst, on="item_id", how="inner")
    .sort_values("user_id")
)
test_df = (
    user_mst.iloc[4000:, :].merge(pref, on="user_id", how="inner").merge(sushi_mst, on="item_id", how="inner")
    .sort_values("user_id")
)
train_df
```

```python
cat_features = range(12)

train_data = train_df.drop(["user_id", "score"], axis=1)
train_label = train_df["score"]
train_group_id = train_df["user_id"]
train_pool = Pool(data=train_data, label=train_label, group_id=train_group_id, cat_features=cat_features)
 
test_data = test_df.drop(["user_id", "score"], axis=1)
test_label = test_df["score"]
test_group_id = test_df["user_id"]
test_pool = Pool(data=test_data, label=test_label, group_id=test_group_id, cat_features=cat_features)
 
# YetiRankはpairwiseでのランク学習の1つ
param = {'loss_function':'YetiRank', 'learning_rate': .05, 'iterations': 200,
         'depth': 4, 'use_best_model':True}
model = CatBoost(param)
model.fit(train_pool, eval_set=test_pool)
```

```python
pred_score = model.predict(test_data)
pred_score
```

```python
def adopt_label_for_lbg(label):
    if (label == 0) | (label == 4): return 1;
    else: return 0

train_data = train_df.drop(["user_id", "score"], axis=1)
train_label = train_df["score"]
train_group = pd.Series([100] * int(train_data.shape[0]/100))

test_data = test_df.drop(["user_id", "score"], axis=1)
test_label = test_df["score"]
test_group = pd.Series([100] * int(test_data.shape[0]/100))

lgb_train_label = train_label.apply(adopt_label_for_lbg)
lgb_test_label = test_label.apply(adopt_label_for_lbg)
```

```python
train_data.columns = [str(i) for i in range(len(train_data.columns))]
test_data.columns = [str(i) for i in range(len(test_data.columns))]
```

```python
train_data
```

```python
uncategorical_feature = [2, 14, 15, 16, 17]
categorical_feature = [i for i in range(18) if not i in uncategorical_feature]
categorical_feature
```

```python
lgb_train = lgb.Dataset(train_data, label = lgb_train_label, weight = train_label, group=train_group)
lgb_vaid =  lgb.Dataset(test_data, label = lgb_test_label, weight = test_label, group=test_group)

max_position = 20
lgbm_params =  {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'lambdarank',
    'metric': 'ndcg',   # for lambdarank
    'ndcg_eval_at': [1,2,3],  # for lambdarank
    'max_position': max_position,  # for lambdarank
    'learning_rate': 1e-8,
    'min_data': 1,
    'min_data_in_bin': 1,
}

uncategorical_feature = [2, 14, 15, 16, 17]
lgb_clf = lgb.train(
    lgbm_params,
    lgb_train,
    categorical_feature= [i for i in range(18) if not i in uncategorical_feature],
    num_boost_round=10,
    valid_sets=[lgb_train, lgb_vaid],
    valid_names=['train','valid'],
    early_stopping_rounds=2,
    verbose_eval=1
)
```
